```python
f, ax = plt.subplots(figsize=(20, 9))
sns.boxplot(data=df)
```

```python
# Boxplot or whiskers plot
keys = len(df.keys())
f, axs = plt.subplots(1,keys, figsize=(20, 9))
for i, key in enumerate(df.keys()):
    sns.boxplot(y=df[key], ax=axs[i])
```

```python
# KDE kernel plot
f, axs = plt.subplots(keys, 1, figsize=(10, 20))
for i, key in enumerate(df.keys()):
    sns.kdeplot(data=df[key], ax=axs[i], fill=True)
    axs[i].set_ylabel(key)
    axs[i].set_xlabel("")
```
